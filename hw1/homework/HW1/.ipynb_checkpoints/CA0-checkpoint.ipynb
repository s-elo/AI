## Using Pickle
You may notice that reading data from the MNIST binary files is time-consuming. A better way is to read from MNIST binary files only once, save the data into pickle files and read from pickle files later.
 
`pickle` is a module used in serializing and de-serializing Python objects. Implement the function `load_and_pickle` and `load_from_pickle` in `assignment/data/mnist.py`.

````python
# Load the data and save into a pickle file
from data.mnist import load_and_pickle
load_and_pickle(mnist_dir)
````

````python
# Load the data from pickle file
from data.mnist import load_from_pickle
data = load_from_pickle(os.path.join(mnist_dir + 'mnist.pkl'))  # Faster, right?

trainX, trainy = data['trainX'], data['trainy']
testX, testy = data['testX'], data['testy']
````

## Flatten the data
Usually, we would expect a $N\times M$ format for data features, where $N$ is the number of data points and $M$ is the number of features. Please code below to flatten the image from $N \times 28 \times 28$ to $N \times 784$.

````python
print("(Before) train image data shape:", trainX.shape)
print("(Before) test image data shape:", testX.shape)
#####################################################################
# TODO:                                                             #
# Change the shape of the data in place                             #
# from N x 28 x 28 to N x 784                                       #
#####################################################################
trainX = trainX.reshape(len(trainX), -1)
testX = testX.reshape(len(testX), -1)
#####################################################################
#                       END OF YOUR CODE                            #
#####################################################################
print("(After) train image data shape:", trainX.shape)
print("(After) test image data shape:", testX.shape)
````

Falttening the data seems pretty naive because we will to some extent lose information (such as spatial correlation between pixels). But still, it is not bad for classification. The visualization below validates that. We can see a clear distinction between different digits.

````python
plt.figure(figsize=(15, 10))
plt.imshow(np.vstack([trainX[trainy == 4][:3000],
                      trainX[trainy == 7][:3000],
                      trainX[trainy == 2][:3000]]).T)
plt.show()
````

When fitting a linear classifier, we often have a bias term. Usually, we prepend one feature to the data (values are all 1 for this feature) to better utilize vectorized operations. Please implement that (prepend a feature).

````python
print("(Before) train image data shape:", trainX.shape)
print("(Before) test image data shape:", testX.shape)
#####################################################################
# TODO:                                                             #
# Change the shape of the data in place                             #
# from N x 28 x 28 to N x 784                                       #
#####################################################################
trainX = np.hstack([np.ones((trainX.shape[0], 1)), trainX])
testX = np.hstack([np.ones((testX.shape[0], 1)), testX])
#####################################################################
#                       END OF YOUR CODE                            #
#####################################################################
print("(After) train image data shape:", trainX.shape)
print("(After) test image data shape:", testX.shape)

if (trainX[:, 0] == 1).all() and (testX[:, 0] == 1).all():
    print("Correct!")
else:
    print("Oops, wrong!")
````

Sometimes in image preprocessing, we would like to subtract the average pixel value computed from all training samples. Please implement that and re-plot the above "average" images.

````python
#####################################################################
# TODO:                                                             #
# Plot the "average" images of digits 1~9 in one figure             #
# (this time with average pixel values subtracted)                  #
#####################################################################
avg_pixel = data['trainX'].mean(axis=0)

plt.figure(figsize=(10, 20))
for i in range(1, 10):
    plt.subplot("19%d" % i)
    plt.imshow((data['trainX'][trainy == i] - avg_pixel).mean(axis=0))
    plt.axis('off')
plt.show()
#####################################################################
#                       END OF YOUR CODE                            #
#####################################################################
````
